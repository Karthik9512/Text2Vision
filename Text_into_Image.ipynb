```python
from diffusers import StableDiffusionPipeline
import matplotlib.pyplot as plt
import torch

artistic_model = "dreamlike-art/dreamlike-diffusion-1.0"

pipe = StableDiffusionPipeline.from_pretrained(
    artistic_model,
    use_safetensors=True
)

prompt = (
    "A happy dog roaming freely in a vast open green field, "
    "golden sunlight, gentle breeze moving the grass, "
    "realistic fur texture, natural colors, shallow depth of field, "
    "cinematic photography, ultra-detailed, 8k resolution"
)

print("Generating image...")

image = pipe(prompt).images[0]

plt.figure(figsize=(6, 6))
plt.imshow(image)
plt.axis("off")
plt.show()
```
